```python
import os
import fiftyone as fo
import fiftyone.zoo as foz

# Where everything will be stored on Google Drive
ROOT = "/content/drive/MyDrive/yolo26_finetune"
EXPORT_DIR = os.path.join(ROOT, "openimages_yolo")
os.makedirs(ROOT, exist_ok=True)

# Classes for your project
OI_CLASSES = ["Dog", "Person"]

# ---- TRAIN SPLIT ----
train_ds = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    classes=OI_CLASSES,
    only_matching=True,
    shuffle=True,
    seed=42,
    max_samples=1500,
)

# YOLO export requires a common class list
classes = train_ds.default_classes

train_ds.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="train",
    classes=classes,
)

# ---- VAL SPLIT ----
val_ds = foz.load_zoo_dataset(
    "open-images-v7",
    split="validation",
    label_types=["detections"],
    classes=OI_CLASSES,
    only_matching=True,
    shuffle=True,
    seed=42,
    max_samples=300,
)

val_ds.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="val",
    classes=classes,
)

print("Exported dataset to:", EXPORT_DIR)
print("Classes:", classes)
print("YAML:", os.path.join(EXPORT_DIR, "dataset.yaml"))
```

```python
yaml_path = os.path.join(EXPORT_DIR, "dataset.yaml")
with open(yaml_path, "r") as f:
    print(f.read())
```

```python
from ultralytics import YOLO
import os

ROOT = "/content/drive/MyDrive/yolo26_finetune"
EXPORT_DIR = os.path.join(ROOT, "openimages_yolo")
yaml_path = os.path.join(EXPORT_DIR, "dataset.yaml")

model = YOLO("yolo26s.pt")

model.train(
    data=yaml_path,
    epochs=40,
    imgsz=640,
    device=0,
    batch=16,
    project=os.path.join(ROOT, "runs"),
    name="yolo26s_openimages_dog_person_v1",
)
```

```python
from ultralytics import YOLO
import os

ROOT = "/content/drive/MyDrive/yolo26_finetune"
best_path = os.path.join(
    ROOT,
    "runs",
    "yolo26s_openimages_dog_person_v1",
    "weights",
    "best.pt",
)

best_model = YOLO(best_path)
metrics = best_model.val(data=os.path.join(ROOT, "openimages_yolo", "dataset.yaml"))
print(metrics)
```
