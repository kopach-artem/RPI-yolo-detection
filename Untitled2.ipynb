```python
import os
import shutil
import fiftyone as fo
import fiftyone.zoo as foz

ROOT = "/content/gdrive/MyDrive/yolo26_finetune"
EXPORT_DIR = os.path.join(ROOT, "openimages_yolo_fast")
os.makedirs(ROOT, exist_ok=True)

# clean previous export to avoid merge / corruption
if os.path.exists(EXPORT_DIR):
    shutil.rmtree(EXPORT_DIR)

# FAST broad subset
TRAIN_SAMPLES = 2000
VAL_SAMPLES = 300

train_ds = foz.load_zoo_dataset(
    "open-images-v7",
    split="train",
    label_types=["detections"],
    shuffle=True,
    seed=42,
    max_samples=TRAIN_SAMPLES,
)

val_ds = foz.load_zoo_dataset(
    "open-images-v7",
    split="validation",
    label_types=["detections"],
    shuffle=True,
    seed=42,
    max_samples=VAL_SAMPLES,
)

classes = train_ds.default_classes

train_ds.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="train",
    classes=classes,
)

val_ds.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="val",
    classes=classes,
)

print("Exported dataset to:", EXPORT_DIR)
print("Number of classes in subset:", len(classes))
print("First 30 classes:", classes[:30])
print("YAML:", os.path.join(EXPORT_DIR, "dataset.yaml"))
```

```python
from ultralytics import YOLO
import os
import torch

ROOT = "/content/gdrive/MyDrive/yolo26_finetune"
EXPORT_DIR = os.path.join(ROOT, "openimages_yolo_fast")
yaml_path = os.path.join(EXPORT_DIR, "dataset.yaml")

model = YOLO("yolo26s.pt")

model.train(
    data=yaml_path,
    epochs=40,
    imgsz=640,
    device=0 if torch.cuda.is_available() else "cpu",
    batch=16 if torch.cuda.is_available() else 8,
    project=os.path.join(ROOT, "runs"),
    name="yolo26s_openimages_fast_v1",
)
```
